### 리그오브레전드 챔피언 정보 데이터 load 및 pickle파일로 저장
- 숫자로 표현된 챔피언 정보를 문자열로 변환할 때 사용

```python
# 챔피언 정보가 있는 url에 request -> response
url = "http://ddragon.leagueoflegends.com/cdn/11.16.1/data/en_US/champion.json"

response = requests.get(url)
response
```

```python
# 필요한 정보만 추출 후 dictionary형태로 저장
champion_datas = {}
for data in response.json()['data']:
    name = response.json()['data'][data]['id']
    key = response.json()['data'][data]['key']

    champion_datas.update({key: name})
```

```python
# 챔피언 정보 pickle 파일로 저장
with open('datas/champion_filter.pkl', 'wb') as f:
    pickle.dump(champion_datas, f)
```

# 2차 데이터 전처리 
- 분석을 위한 데이터셋 세분화

### 한 게임당 10명의 유저 경기 데이터를 보유하고 있는 데이터셋
- 한 게임에 10명의 유저데이터가 있기 때문에 1row(한 경기) -> 10rows(한 유저) 데이터로 세분화

```python
participant_list = []
for num in range(len(df)):
    one_game_data = df.loc[num]
    for i in range(10):
        row = {k:v for k, v in one_game_data['participants'][i]['stats'].items() if k != "timeline"}
        row['role'] = one_game_data['participants'][i]['timeline']['role']
        row['championid'] = one_game_data['participants'][i]['championid']
        row['spell1id'] = one_game_data['participants'][i]['spell1id']
        row['spell2id'] = one_game_data['participants'][i]['spell2id']

        participant_list.append(row)
    
participants_df = pd.DataFrame(participant_list)
```

```python
participants_df.tail()
```

### 한 게임당 10명의 유저 라인 및 티어 데이터를 보유하고 있는 데이터셋
- 한 게임에 10명의 유저데이터가 있기 때문에 1row(한 경기) -> 10rows(한 유저) 데이터로 세분화

```python
participantextendedstat_list = []
for num in range(len(df)):
    one_game_data = df.loc[num]
    for i in range(10):
        row = {k:v for k, v in one_game_data['participantextendedstats'][i].items() if k != "participantid"}
        participantextendedstat_list.append(row)
    
participantextendedstat_df = pd.DataFrame(participantextendedstat_list)
```

```python
participantextendedstat_df.tail()
```

### 한 게임당 두 팀의 오브젝트 데이터를 보유하고 있는 데이터셋
- 1. 1row = 유저인 데이터셋
    - 위 두 데이터셋과 동일하게 row를 유저데이터로 맞추기 위해 1row -> 10rows 로 세분화
    - 분석을 위해 gameduration 컬럼데이터 추가
- 2. 1row = 팀인 데이터셋
    - 오브젝트 중요도에 대한 분석을 위한 데이터셋
    - 1row(한 경기) -> 2rows(한 팀) 으로 세분화

#### 1. 1row = 유저인 데이터셋

```python
team_list = []
for num in range(len(df)):
    one_game_data = df.loc[num]
    for i in range(10):

        if i < 5:
            row = {k:v for k, v in one_game_data['teams'][0].items() if k != "bans"}
            row['bans'] = (one_game_data['teams'][0]['bans'][i]['championid'])
        else:
            row = {k:v for k, v in one_game_data['teams'][1].items() if k != "bans"}
            row['bans'] = (one_game_data['teams'][1]['bans'][i-5]['championid'])
        
        row['gameduration'] = one_game_data['gameduration']
        
        team_list.append(row)
        
team_df = pd.DataFrame(team_list)
```

```python
team_df.tail()
```

#### 2. 1row = 팀인 데이터셋
- 조기서렌 게임 삭제

```python
team_list = []
for num in range(len(df)):
    one_data = df.loc[num]
    for i in range(len(one_data['teams'])):
        if one_data['gameduration'] <= 605: # 결측치 처리 과정에서 605초 이전 데이터를 삭제하면 비정상적인 데이터가 다 삭제되는 것을 확인
            continue
        row = {k:v for k, v in one_data['teams'][i].items() if k != "bans"}
        row['gameduration'] = one_data['gameduration']
        team_list.append(row)
        
team_df_2 = pd.DataFrame(team_list)
```
